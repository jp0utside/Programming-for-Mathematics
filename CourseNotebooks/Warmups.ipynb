# Warmup 3/2
Central limit theorem: As N -> infinity, the distribution SQRT(N) * (means - mu) approaches N (0, sigma^2)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
C = np.array([1,1,1,1,6])
mu = np.mean(C)
sigma = np.std(C)
print(mu, sigma)
```

```python
N = 50
samp = np.random.choice(C, N)
print(samp)
print(np.mean(samp))

tests = np.random.choice(C, (1000,N))
means = np.mean(tests, axis = 1)
means.shape
print(means[:10])
plt.hist(means)
plt.show()
```

```python
devs = np.sqrt(N) * (means - mu)
x = np.linspace(-6, 6, 1000)
y = (1/(sigma*np.sqrt(2*np.pi))) * np.exp(-0.5*(x/sigma)**2)
plt.figure(figsize = (6,4))
plt.hist(devs, bins = 20, density = True)
plt.plot(x, y)
plt.title("Central Limit Theorem")
plt.show()
```

#Warmup 3/9: Imaging



*   **Complex Numbers**
  *   C = R + Rj (Where C is set of complex numbers, R is reals, j is complex imaginary number)
  *   In python, imaginary numbers represented as "j"
*   **Mandelbrot Sets**
  *   Let c be a complex number
  *   Comsider the sequence: 0, c, c^2 + c, (c^2 + c)^2 + c, ...
  *   Eventually value blows up
  *   Madelbrot set is all c's such that the sequence remains bounded
  *   Graph shows point if for (x,y), c = x + yj doesn't blow up
*   **Process**
  *   Choose a big circle, and a point inside
  *   apply function
  *   if point stays inside circle, it is valid



```python
import numpy as np
from PIL import Image
```

```python
def stays_bounded(c):
  sum = c
  for i in range(10):
    sum = sum**2 + c
    if sum > 2:
      return False
  return True

stays_bounded(-2)
```

```python
re_array = np.linspace(-2, 2, 500)
im_array = np.linspace(-1, 1, 1000)
c_array = np.add.outer(re_array, im_array * 1j)

def make_mandel(num_iter = 100):
  arr = np.zeros(c_array.shape)
  b_arr = np.zeros(c_array.shape)
  for t in range(num_iter):
    arr = arr**2 + c_array
    arr_size = np.abs(arr)
    is_bded = (arr_size <= 2)
    b_arr = b_arr + is_bded
    arr = (arr * is_bded) + (2*(1-is_bded))
  return b_arr, is_bded
```

```python
esc_time, mbrot = make_mandel()
pixels = np.zeros((c_array.shape[0], c_array.shape[1], 3))
pixels[:,:,0] = mbrot * 255
pixels[:,:,1] = mbrot * 255
pixels[:,:,2] = mbrot * 255
#pixels[:,:,0] += (2*esc_time % 255)
pixels[:,:,1] += (200*esc_time % 255)
pixels[:,:,2] += (37*esc_time % 255)
#pixels = np.clip(pixels, 0, 255)
img = Image.fromarray(pixels.astype(np.uint8))
display(img)
```
